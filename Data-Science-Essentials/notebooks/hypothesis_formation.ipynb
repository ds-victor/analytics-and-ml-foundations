# Topic 4: Exploratory Hypothesis Formation

In this notebook, we move from simple plotting to forming and testing hypotheses about our data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

df = pd.read_csv(Path("..") / "data" / "raw" / "real_estate_dataset.csv")

## Hypothesis 1: Larger square footage leads to higher prices.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='area_sqft', y='price', hue='region')
plt.title("Price vs. Square Footage by Region")
plt.show()
```

```python
## Hypothesis 2: Newer houses (age < 15 years) are more expensive than older ones.
df['is_new'] = df['age_years'] < 15
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='is_new', y='price')
plt.title("Price Comparison: Newer vs. Older Houses")
plt.show()
```
