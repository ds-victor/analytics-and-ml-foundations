# Topic 2: Jupyter & IDE Setup - House Price EDA

This notebook demonstrates the initial setup and basic EDA to ensure our environment is correctly configured.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

# Set plotting style
sns.set_theme(style="whitegrid")

# Load data using relative paths
DATA_PATH = Path("..") / "data" / "raw" / "real_estate_dataset.csv"
df = pd.read_csv(DATA_PATH)

print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# Quick visualization of price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title("Distribution of House Prices")
plt.show()
```
