# Business Analytics – BI Preparation Notebook

Aggregate data for dashboard building.

```python
import pandas as pd

df = pd.read_csv('../data/raw/real_estate_dataset.csv')

# Implement variable cost ratio based on property type
cost_ratio = {'villa': 0.80, 'apartment': 0.60, 'townhouse': 0.70, 'house': 0.75, 'condo': 0.65}
df['cost'] = df['price'] * df['property_type'].map(cost_ratio)
df['estimated_roi'] = ((df['price'] - df['cost']) / df['cost'] * 100).round(2)

agg = df.groupby(['property_type', 'year_sold']).agg(
    avg_price=('price', 'mean'),
    median_price=('price', 'median'),
    total_sales=('id', 'count'),
    avg_days_on_market=('days_on_market', 'mean'),
    avg_roi=('estimated_roi', 'mean')
).reset_index()

agg.to_csv('../data/processed/business_aggregates.csv', index=False)
print('Aggregated data with variable ROI saved for dashboard building.')
```
