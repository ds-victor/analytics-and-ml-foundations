# Topic 5: Feature Engineering Exploration

Testing new feature ideas before implementing them in our production scripts.

```python
import pandas as pd
import numpy as np
from pathlib import Path

df = pd.read_csv(Path("..") / "data" / "raw" / "real_estate_dataset.csv")

# 1. Create Price per Sqft
df['price_per_sqft'] = df['price'] / df['area_sqft']

# 2. Property Age Group
df['age_group'] = pd.cut(df['age_years'], bins=[0, 10, 30, 100], labels=['New', 'Mid', 'Old'])

# 3. Log transformation for Price
df['log_price'] = np.log1p(df['price'])

print("New Features Preview:")
df[['price', 'log_price', 'price_per_sqft', 'age_years', 'age_group']].head()
```
