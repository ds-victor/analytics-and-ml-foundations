# Topic 9: Fairness & Ethics Analysis

Visualizing disparate impact and bias in our real estate model.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

df = pd.read_csv(Path("..") / "data" / "raw" / "real_estate_dataset.csv")

plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='region', y='price')
plt.title("Price Distribution Across Regions (Fairness Check)")
plt.show()
```
